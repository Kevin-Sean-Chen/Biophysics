```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import time
import math

% matplotlib inline
```

## Multiplicative and Additive noise

```python
L = 1000
X = np.sin(np.vstack((np.arange(0,L),np.arange(0,L)))/10)  #signel
gn = np.random.randn(2,L)  #multiplicative noise
nn = np.random.randn(2,L)  #additive noise
```

```python
#signal
yn = X  #measurement
plt.plot(yn.T);
```

```python
#covariance measure for two signals
plt.imshow(np.cov(yn),interpolation='None')
plt.colorbar()
```

```python
#additive
yn = X + nn  #measurement
plt.plot(yn.T);
```

```python
plt.imshow(np.cov(yn),interpolation='None')
plt.colorbar()
```

```python
#difference
plt.hist(np.diff(yn,axis=0).T,bins=100);
```

```python
#with multiplicative noise
yn = np.exp(gn)*X + nn  #measurement
plt.plot(yn.T);
```

```python
plt.imshow(np.cov(yn),interpolation='None')
plt.colorbar()
```

```python
plt.hist(np.diff(yn,axis=0).T,bins=100);
```

### Ans:

Given that the distribution of difference between $y_1$ and $y_2$ are non-Gaussian under multiplicative noise, as shown above, one can potentially recover the signal by exluding "outliers" due to random noise in the exponent. We can fit the histogram shown above with mixture model with Gaussian with the mean of an exponential of Gaussian noise, as known from the equations. Fitting the log-likelihood would be a helpful way to separate the variables for fitting.

```python
#distribution of exponential noise
plt.hist(np.exp(np.random.randn(10000)),bins=100);
```
